```python
files = sorted(glob('*textile*txt*'))[2::3]

fig, ax = plt.subplots(1,1,figsize=(15,10))

power = 0.00530899  #0.00526478  # in watts
d = 0.59            # mm
d = d/1000          # m
A = math.pi*((d/2)**2)
I = power/A

for file in files:
    df = pd.read_csv(file, sep='\t', skiprows=8)
    
    t = df['#Time']
    dE = df['dE2000']
    H = (I * t) / 1e6
    
    ax.plot(H,dE)
    
```
